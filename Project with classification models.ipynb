###  We plotted the collinearity between all the variables and dropped the features with an absolute correlated coefficient value of above 0.8. Body Mass Index seem to be highly correlated with other features and it was dropped.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize= (10, 10))
sns.heatmap(features.corr())
```

```python
import numpy as np

# Create correlation matrix
corr_matrix = features.corr().abs()

# Select upper triangle of correlation matrix 
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool)) 

# Find index of feature columns with correlation greater than 0.8 
to_drop = [column for column in upper.columns if any(upper[column] > 0.8)] 

new_features = features.drop(to_drop, axis=1)
```

###   We also converted the target variable, Absenteeism time in hours, to categorical variable for our initial analysis. The following are the thresholds we used to convert our target variable to categorical variable.


Target Value Threshold by Groups

Group Number	Threshold

0	Number of hours=0

1	0 < Number of hours <= 6

2	Number of hours > 6


```python
Absenteeism2 = Absenteeism
g1 = (Absenteeism2['Absenteeism time in hours']==0)
g2 = (Absenteeism2['Absenteeism time in hours']>0) & (Absenteeism2['Absenteeism time in hours']<=6)
g3 = (Absenteeism2['Absenteeism time in hours']>6)
```

```python
Absenteeism2.loc[g1,'Absenteeism time in hours'] = 0
Absenteeism2.loc[g2,'Absenteeism time in hours'] = 1
Absenteeism2.loc[g3,'Absenteeism time in hours'] = 2
```

```python
Absenteeism2['Absenteeism time in hours'].value_counts()
```

```python
target = Absenteeism2['Absenteeism time in hours']
```

##  Data splitting

```python
# Splitting the model
from sklearn.model_selection import train_test_split
y = target
X = Absenteeism2.drop(['Absenteeism time in hours'],axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,
random_state=12345)
```

```python
# Random forest technique
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
y_predict = clf.predict(X_test)
print(confusion_matrix(y_test, y_predict))
print("Accuracy of Random Forest Technique is:", accuracy_score(y_test, y_predict)*100)
```

```python
# Decision Tree technique
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
dt_model = tree.DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
y_predict = dt_model.predict(X_test)
print(confusion_matrix(y_test, y_predict))
print("Accuracy of Decision Tree Technique is:", accuracy_score(y_test, y_predict)*100)
```

```python
from sklearn.tree import export_graphviz

# export the decision tree to a tree.dot file
# for visualizing the plot easily anywhere
export_graphviz(dt_model, out_file ='tree.dot', feature_names =X.columns)
```

```python
# NaÃ¯ve Bayes Classifiers
from sklearn.naive_bayes import GaussianNB 
gnb = GaussianNB() 
gnb = gnb.fit(X_train, y_train)
y_predict = gnb.predict(X_test)
print(confusion_matrix(y_test, y_predict))
print("Accuracy of Naive Bayes Classifiers is:", accuracy_score(y_test, y_predict)*100)
```

```python
# Support Vector Machine
# Normalizing the data & training the Normalized data
from sklearn import preprocessing
y = target
X = Absenteeism2.drop(['Absenteeism time in hours'],axis=1)
Normalized_X = preprocessing.normalize(X)
X_train, X_test, y_train, y_test = train_test_split(Normalized_X, y, test_size=0.20,
random_state=12345)
```

```python
# Design the model
from sklearn import svm
support = svm.SVC(kernel='linear')
support.fit(X_train, y_train)
y_predict = support.predict(X_test)
print(confusion_matrix(y_test, y_predict))
print("Accuracy of Support Vector Machine through linear kernel is:", accuracy_score(y_test, y_predict)*100)
```

```python
#ROC curve on decision tree and random forest
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
y = target
rf_clf = RandomForestClassifier(max_depth= 5, n_estimators=60)
y_score = rf_clf.fit(X_train, y_train).predict_proba(X_test)
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])
n_classes = y_test_bin.shape[1]
# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
color_dict = ['darkorange', 'lightseagreen', 'coral']
label_dict = ['Absenteeism time equals 0', 'Absenteeism time between 0 and 6', 'Absenteeism time morethan 6']

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    plt.plot(fpr[i], tpr[i], color=color_dict[i], lw=2, label =label_dict[i] )
    print('AUC for Class {}: {}'.format(i+1, auc(fpr[i], tpr[i])))
    
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Curves with Random Forest')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
